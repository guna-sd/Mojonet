```python
import subprocess
import time
from matplotlib import pyplot as plt

def run_command(command):
    start_time = time.time()
    result = subprocess.run(command, shell=True)
    end_time = time.time()
    return end_time - start_time

# Define commands to run matrix multiplication implementations in different languages
commands = {
    "Python": "python3 /home/guna/bench/mat.py",  # Python implementation path
    "C": "/home/guna/bench/cmat",                 # C implementation path
    "C++": "/home/guna/bench/cppmat",             # C++ implementation path
    "Rust": "/home/guna/bench/mat/target/debug/mat",  # Rust implementation path
    "Mojo": "/home/guna/bench/mojomat"            # Mojo implementation path
}

# Run each implementation and measure execution time
timings = {}
for lang, cmd in commands.items():
    timing = []
    runs = 1 if lang == "Python" else 5  # Run Python implementation once, others 5 times
    for _ in range(runs):
        timing.append(run_command(cmd))
    timings[lang] = sum(timing) / len(timing)

# Print timings and plot the results
for lang, time in timings.items():
    print(f"{lang}: {time} seconds")

plt.bar(timings.keys(), timings.values())
plt.ylabel('Time (seconds)')
plt.title('Matrix Multiplication Speed Comparison')
plt.xticks(rotation=45)
plt.tight_layout()  # Adjust layout to make room for the rotated x-axis labels
plt.show()
```
